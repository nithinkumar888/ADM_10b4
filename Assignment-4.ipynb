```python
  import pandas as pd
data = pd.read_csv('/content/Titanic-Dataset.csv')

numerical_data = data.select_dtypes(include=['number'])

mean_values = numerical_data.mean()
median_values = numerical_data.median()
mode_values = numerical_data.mode().iloc[0]

# Print the results
print("Mean values for numerical columns:\n", mean_values)
print("\nMedian values for numerical columns:\n", median_values)
print("\nMode values for numerical columns:\n", mode_values)
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Age'].fillna(data['Age'].mode(), inplace=True)
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_normilalize = ['Age', 'Fare']
data[columns_to_normilalize] = scaler.fit_transform(data[columns_to_normilalize])
data[columns_to_normilalize].head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_standardize = ['Age', 'Fare']
data[columns_to_standardize] = scaler.fit_transform(data[columns_to_standardize])
data[columns_to_standardize].head()
```
